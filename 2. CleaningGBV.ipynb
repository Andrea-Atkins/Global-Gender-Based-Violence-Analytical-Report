```python
who_df = pd.read_csv('Gender-Based-Violence/Output/who_ipv_clean.csv')
unodc_df = pd.read_csv('Gender-Based-Violence/Output/unodc_homicide_clean.csv')
legal_df = pd.read_csv('Gender-Based-Violence/Output/legal_protection_clean.csv')
oda_df = pd.read_csv('Gender-Based-Violence/Output/oda_clean.csv')

print(who_df.shape, unodc_df.shape, legal_df.shape, oda_df.shape)
```

```python
master_df = who_df.merge(unodc_df, on='iso3', how='outer', suffixes=('', '_unodc'))
master_df = master_df.merge(legal_df, on='iso3', how='outer', suffixes=('', '_legal'))
master_df = master_df.merge(oda_df, on='iso3', how='outer', suffixes=('', '_oda'))

print(master_df.shape)
print(master_df.columns.tolist())
```

```python
master_df['country'] = master_df['country'].combine_first(master_df['country_unodc']).combine_first(master_df['country_legal']).combine_first(master_df['country_oda'])
master_df['region'] = master_df['region'].combine_first(master_df['region_legal'])

master_df = master_df.drop(columns=['country_unodc', 'country_legal', 'region_legal', 'country_oda'])
print(master_df.shape)
print(master_df.columns.tolist())
```

```python
def normalize(series):
    filled = series.fillna(series.median())
    return (filled - filled.min()) / (filled.max() - filled.min()) * 100

master_df['ipv_norm'] = normalize(master_df['ipv_prevalence'])
master_df['homicide_norm'] = normalize(master_df['female_homicide_rate'])
master_df['legal_norm_inverted'] = 100 - normalize(master_df['safety_legal_score'])  # invert: low legal score = high risk
master_df['funding_norm_inverted'] = 100 - normalize(master_df['oda_received_usd'])  # invert: low funding = high risk

master_df.head()
```

```python
# Step 1: Require at least one real severity measure (IPV or homicide) to be included at all
master_df_filtered = master_df[
    master_df['ipv_prevalence'].notna() | master_df['female_homicide_rate'].notna()
].copy()

print("Countries with at least one severity measure:", master_df_filtered.shape)

# Step 2: Re-normalize using only this filtered set
master_df_filtered['ipv_norm'] = normalize(master_df_filtered['ipv_prevalence'])
master_df_filtered['homicide_norm'] = normalize(master_df_filtered['female_homicide_rate'])
master_df_filtered['legal_norm_inverted'] = 100 - normalize(master_df_filtered['safety_legal_score'])
master_df_filtered['funding_norm_inverted'] = 100 - normalize(master_df_filtered['oda_received_usd'])

master_df_filtered['vulnerability_index'] = (
    master_df_filtered['ipv_norm'] * 0.40 +
    master_df_filtered['homicide_norm'] * 0.25 +
    master_df_filtered['legal_norm_inverted'] * 0.20 +
    master_df_filtered['funding_norm_inverted'] * 0.15
)

master_df_filtered = master_df_filtered.sort_values('vulnerability_index', ascending=False).reset_index(drop=True)
master_df_filtered.head(20)
```

```python
master_df_filtered.to_csv('Gender-Based-Violence/Output/master_vulnerability_index.csv', index=False)

top20 = master_df_filtered[['iso3', 'country', 'region', 'income_group', 'vulnerability_index']].head(20)
top20.to_csv('Gender-Based-Violence/Output/question5_top20_underserved.csv', index=False)

print(os.listdir('Gender-Based-Violence/Output'))
```
